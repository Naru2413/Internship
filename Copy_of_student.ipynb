```python
from google.colab import files
uploaded = files.upload()

# âœ… Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# âœ… Load the dataset (note the sep=';' here!)
df = pd.read_csv('student-mat.csv', sep=';')

# âœ… Clean column names (just in case)
df.columns = df.columns.str.strip()

# âœ… Debug: print all columns
print("\nðŸ“Œ Column Names:")
print(df.columns.tolist())

# âœ… Show the first few rows
print("\nðŸ“Œ First 5 rows:")
print(df.head())

# âœ… Check for missing values
print("\nðŸ“Œ Missing values:")
print(df.isnull().sum())

# âœ… Check data types
print("\nðŸ“Œ Data types:")
print(df.dtypes)

# âœ… Dataset shape
print("\nðŸ“Œ Dataset shape:")
print(df.shape)

# âœ… Remove duplicates
df = df.drop_duplicates()

# âœ… 1. Average G3 score
average_g3 = df['G3'].mean()
print("\nðŸŽ¯ Average G3 score:", average_g3)

# âœ… 2. Students scoring above 15 in G3
above_15 = df[df['G3'] > 15].shape[0]
print("ðŸŽ¯ Students scoring above 15:", above_15)

# âœ… 3. Correlation between studytime and G3
correlation = df['studytime'].corr(df['G3'])
print("ðŸŽ¯ Correlation between study time and G3:", correlation)

# âœ… 4. Average G3 by gender
gender_avg = df.groupby('sex')['G3'].mean()
print("ðŸŽ¯ Average G3 by gender:\n", gender_avg)

# âœ… 5. Histogram of G3
plt.figure(figsize=(8, 5))
plt.hist(df['G3'], bins=10, color='skyblue', edgecolor='black')
plt.title("Distribution of Final Grades (G3)")
plt.xlabel("Final Grade (G3)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# âœ… 6. Scatter: Study time vs G3
plt.figure(figsize=(8, 5))
plt.scatter(df['studytime'], df['G3'], color='green', alpha=0.6)
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.grid(True)
plt.show()

# âœ… 7. Bar chart: Average G3 by gender
plt.figure(figsize=(6, 4))
sns.barplot(x=gender_avg.index, y=gender_avg.values, palette='pastel')
plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average G3")
plt.ylim(0, 20)
plt.grid(axis='y')
plt.show()
```
